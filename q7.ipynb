```python
im11 = np.mean(imread("11.jpg"), axis=2)

im12 = np.mean(imread("12.jpg"), axis=2)
```

```python
pylab.imshow(im12,cmap='gray'),pylab.axis('off')
```

```python
from scipy import signal
freq12 = fp.fft2(im12)
gauss_kernel1 = np.outer(signal.gaussian(im12.shape[0],2),signal.gaussian(im12.shape[1],2))
gauss_kernel1.shape[1]
freq_kernel = fp.fft2(fp.ifftshift(gauss_kernel1))
convolved = freq12*freq_kernel
im12_= fp.ifft2(convolved).real
```

```python
imshow(im12_,cmap='gray')
```

```python
(w,h)=freq12.shape
half_w, half_h = int(w/2), int(h/2)
freq12Shifted=fp.fftshift(freq12)
freq2_low = np.copy(freq12Shifted)
freq2LP=np.copy(freq12Shifted)
freq2_low[half_w-100:half_w+100,half_h-100:half_h+100] = 0 
freq2LP -= freq2_low 
im1 = fp.ifft2(fftpack.ifftshift(freq2LP)).real
pylab.imshow(im1, cmap='gray'),pylab.axis('off')
```
