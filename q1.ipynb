```python
im=Image.open('image1.jpg')
```

```python
pylab.figure(figsize=(20,30))
num_colors_list = [1 << n for n in range(7,-1,-1)]
i = 1
for num_colors in num_colors_list:
  im1 = im.convert('P', palette=Image.ADAPTIVE, colors=num_colors)
  pylab.subplot(4,2,i), pylab.imshow(im1), pylab.axis('off')
  pylab.title('Image with # colors = ' + str(num_colors) , size=20)
  i += 1
  pylab.subplots_adjust(wspace=0.2, hspace=0)
  pylab.show()
```

```python
def differencePixels(img,x,y):
  fromtop=img[x][y]-img[x][y-1]
  fromright=img[x][y]-img[x+1][y]
  fromleft=img[x][y]-img[x-1][y]
  frombottom=img[x][y]-img[x][y+1]
  print("difference from neighbor pixels: \n top:{} \n rihgt:{} \n left:{} \n bottom:{}".format(fromtop,fromright,fromleft,frombottom))
```

```python
im2=np.array(img1)
differencePixels(im2,5,10)
```
