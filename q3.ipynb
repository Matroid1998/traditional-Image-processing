```python
from PIL import Image
im1=Image.open("image1.jpg")
```

```python
pylab.subplot(1,2,1), pylab.imshow(im1), pylab.title('Original Image',
size=20), pylab.axis('off')
box_blur_kernel = np.reshape(np.ones(5*5),(5,5)) / (5*5)
im1_blur = im1.filter(ImageFilter.Kernel((5,5), box_blur_kernel.flatten()))
pylab.subplot(1,2,2), pylab.imshow(im1_blur), pylab.title('blur image',
size=20), pylab.axis('off')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)  
ax2 = fig.add_subplot(122) 
im2 = imageio.imread("image2.jpg")
result = ndimage.median_filter(im2, size=5)
ax1.imshow(im2)
ax2.imshow(result)
plt.show()
```
