```python
def sharpenedImage(im,alpha=10,sigma=5):
  im_blurred = ndimage.gaussian_filter(im, sigma)
  im_detail = np.clip(im - im_blurred, 0, 1)
  pylab.gray()
  fig, axes = pylab.subplots(nrows=2, ncols=2, sharex=True, sharey=True,
  figsize=(15, 15))
  axes = axes.ravel()
  axes[0].set_title('Original image', size=15), axes[0].imshow(im)
  axes[1].set_title('Blurred image, sigma=5', size=15),
  axes[1].imshow(im_blurred)
  axes[2].set_title('Detail image', size=15), axes[2].imshow(im_detail)
  im_sharp = np.clip(im + alpha*im_detail, 0, 1)
  axes[3].set_title('Sharpened, alpha=10', size=15),
  axes[3].imshow(im_sharp)
  for ax in axes:
      ax.axis('off')
      fig.tight_layout()
      pylab.show()
```
